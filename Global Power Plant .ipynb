As per dataset and columns we viewed and found that we have to drop some columns for proper modelling and columns with less mission values needs to fill the NAN gaps with values by mean or mode method

```python
df_G=df_GP.drop(columns=['country','country_long','name','gppd_idnr','owner','url','other_fuel1','other_fuel2','other_fuel3','wepp_id'],axis=1)
df_G.head()
```

We dropped some of the columns which was not more useful let's check for the other columns and find the patterns in dateset 

As per our observation we can see and found the generation columns have NAN more than 50% so for now we could add the values in one and make some sensical data

```python
df_G['Total Generation']=df_G['generation_gwh_2013']+df_G['generation_gwh_2014']+df_G['generation_gwh_2015']+df_G['generation_gwh_2016']+df_G['generation_gwh_2017']+df_G['generation_gwh_2018']+df_G['generation_gwh_2018']
df_G
```

visualization of the dataset

filling the null values

```python
#replacing the missing the values  in age with mean age 
df_G['commissioning_year'].fillna(df_G['commissioning_year'].mean(), inplace=True)
df_G['year_of_capacity_data'].fillna(df_G['year_of_capacity_data'].mean(), inplace=True)
df_G['latitude'].fillna(df_G['latitude'].mode([0]))
```

```python
df_G=df_G.drop(columns=['generation_gwh_2013','generation_gwh_2014','generation_gwh_2015','generation_gwh_2016','generation_gwh_2017','generation_gwh_2018','generation_gwh_2019'],axis=1)
df_G.head()
```
